```python
import numpy as np
import pandas as pd
import json

#photo data
c_data = pd.read_json('checkin.json',lines=True)
#business data
b_data = pd.read_json('business.json',lines=True)
```

```python
Day_list = []

for i in range(0,len(c_data['time'])):
    a = list(c_data.loc[i,'time'].keys())
    for b in range(0,len(a)):
        c = a[b]
        if c not in Day_list:
            Day_list.append(c)

    
Day_list[0]

for i in Day_list:
    c_data[i+'_f'] = c_data['time'].map(lambda x: 1 if i in x else 0)
```

```python
c_data_1 = pd.DataFrame(
    c_data.iloc[:,1].values.tolist())

c_data_new = pd.concat([c_data,c_data_1], axis=1)

c_data_new.fillna(0)
```
